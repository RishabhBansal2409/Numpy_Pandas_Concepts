# Pandas Data Visualization using Matplotlib

**Importing the required libraries.**

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

**Loading the dataset.**

```python
df3 = pd.read_csv('df3')
```

**Checking the properties of the column present in the dataframe.**

```python
df3.info()
```

**Inspecting the dataset.**

```python
df3.head()
```

**Creating a scatter plot of a vs b.**

```python
df3.plot.scatter('a' , 'b', color = 'red' , s = 50 , figsize = (12,3) , alpha = 1)
```

**Create a histogram of the 'a' column and setting the style to 'ggplot' to style the sheet.**

```python
plt.style.use('ggplot')
```

```python
df3['a'].plot.hist(bins = 50, alpha = 0.5)
```

**Creating a boxplot comparing the a and b columns.**

```python
df3[['a','b']].plot.box()
```

**Creating a kde plot of the 'd' column**

```python
df3['d'].plot.density()
```

**Increasing the linewidth and making the linestyle dashed.**

```python
df3['d'].plot.density(ls = '--' , lw = 3)
```

**Creating an area plot of all the columns for just the rows up to 30.**

```python
df3.iloc[:30,:].plot.area(alpha = 0.3)
```

**Displaying the legend outside of the plot.**

```python
df3.iloc[:30,:].plot.area(alpha = 0.3)
plt.legend(bbox_to_anchor=(1, 0.5))
```
